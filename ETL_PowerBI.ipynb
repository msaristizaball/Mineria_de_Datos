# ETL en Python

```python
# Cargamos las librerías para el análisis de datos
import pandas as pd
```

```python
# Cargamos la base de datos
BD = "/content/Matriz a Tabular.csv" # Copiamos la ruta del archivo
Base = pd.read_csv(BD, sep="\t") # Cargamos la base, indicándole que el separador es tabular 
                                 # (puede cambiar archivo a archivo)
```

```python
# Visualizamos la base de datos
Base

# Para bases de datos muy grandes se recomienda imprimir los primeros registros solamente
# Con Base.head(5) mostramos los primeros 4 registros
```

```python
# Eliminamos la última fila
Base.drop(Base.tail(1).index,axis=0,inplace=True)
```

```python
# Visualizamos la base de datos nuevamente
Base
```

```python
# Visualizamos los nombres de las columnas
Base.columns
```

```python
# Convertimos la primera fila en los nombres de las columnas
Base.columns = Base.iloc[0,:]

# Base.iloc[0,:] toma la primera fila de la base de datos.
# El primer parámetro, 0, indica que es la fila 0
# El segundo parámetro, :, indica que vamos a tomar todas las columnas de la fila
```

```python
# Visualizamos los nombres de las columnas nuevamente
Base.columns
```

```python
# Eliminamos la primera fila
Base.drop(0,axis=0,inplace=True) #Indicamos que vamos a eliminar la fila 0 (la primera), axis=0 indica que estamos eliminando una fila (si fuera axis=1 sería una columna), e inplace=True para hacer la eliminación en la base
Base.reset_index(drop=True, inplace=True) # Reiniciamos los índices para evitar problemas futuros en la indexación. Con drop=True evitamos que los índices se inserten como una columna
```

```python
# Visualizamos la base de datos nuevamente
Base
```

```python
# Transformamos la base de datos de tipo matricial a tipo tabular
FACT_BaseNueva = pd.melt(Base, id_vars=["Producto"], var_name='Fechas', value_name='Valor')
# En el primer parámetro ponemos la base de datos, en el segundo cuál es nuestra columna de ID's, en el tercero y cuarto cómo queremos que se llamen nuestras columnas de variables y valores respectivamente

FACT_BaseNueva = FACT_BaseNueva.dropna() # Eliminamos los valores nulos
print(FACT_BaseNueva) # Visualizamos el dataframe
```

```python
# Exportamos a Excel
FACT_BaseNueva.to_excel("ExcelTabular.xlsx", sheet_name='FACT_BaseNueva', index=False)
```
